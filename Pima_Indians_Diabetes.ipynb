```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('diabetes.csv')

# Убираем Outcome
features = df.drop(columns=['Outcome'])

print("Количество объектов:", len(df))

# Таблица статистик
stats = pd.DataFrame()

# a) количество объектов
stats['count'] = features.count()

# б) размах (max - min)
stats['range'] = features.max() - features.min()

# в) среднее
stats['mean'] = features.mean()

# д) мода
stats['mode'] = features.mode().iloc[0]

# е) медиана
stats['median'] = features.median()

# ф) стандартное отклонение
stats['std'] = features.std()

# п) коэффициент вариации (std / mean)
stats['cv'] = stats['std'] / stats['mean']

print("\nОписательная статистика:\n")
print(stats)

# 🔹 Частотные диаграммы
features.hist(figsize=(12, 8))
plt.suptitle("Гистограммы признаков")
plt.show()

# 🔹 Диаграммы с усами (boxplot)
plt.figure(figsize=(12, 6))
sns.boxplot(data=features)
plt.title("Boxplot признаков")
plt.xticks(rotation=45)
plt.show()
```

**Анализ описательной статистики**

*Среднее значение Pregnancies* составляет около 3.85 при достаточно большом разбросе (std ≈ 3.37), что указывает на высокую вариативность количества беременностей и наличие значений, значительно превышающих среднее.

*Для Glucose* среднее значение равно 120.9 при умеренном стандартном отклонении (≈ 31.97), что говорит о заметном разбросе значений и наличии пациентов с повышенным уровнем глюкозы.

*BloodPressure* имеет среднее значение 69.1 и относительно небольшой коэффициент вариации, что указывает на умеренную изменчивость, однако наличие минимальных значений, близких к нулю, говорит о возможных ошибках в данных.

*В признаках SkinThickness и Insulin* наблюдается высокая вариативность (особенно Insulin, где коэффициент вариации > 1), а также нулевая мода, что подтверждает наличие большого количества пропусков или некорректных значений.

*Признак BMI* имеет среднее значение около 32 при относительно низком разбросе, что указывает на преобладание избыточной массы тела среди пациентов.

*DiabetesPedigreeFunction* характеризуется умеренной вариативностью и в целом невысокими значениями, однако присутствуют отдельные высокие значения.

*Средний возраст пациентов* составляет около 33 лет, при этом разброс значений умер
енный, что говорит о преобладании пациентов молодого и среднего возраста.

*В целом* данные характеризуются наличием разброса, выбросов и возможных пропусков, особенно в признаках Insulin и SkinThickness, что требует дополнительной предобработки и указывает на наличие факторов риска диабета в выборке.
