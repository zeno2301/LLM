# Лаба3

1. Загрузить данные «Pima Indians Diabetes» c ресурса Kaggle.com
2. Произвести расчет влияния 8 параметров на целевой показатель «Outcome».
3. Дополнительно. Использовать для загрузки данных и расчета библиотеки: Scipy.stats

```
import pandas as pd
import numpy as np
from scipy import stats
import os

# Загрузка данных в DataFrame
# Предполагается, что файл diabetes.csv уже находится в директории /content/
data_file_path = "/content/diabetes.csv"

if os.path.exists(data_file_path):
    df = pd.read_csv(data_file_path)
    print("Данные успешно загружены:")
    print(df.head())
    print("\nРазмер данных:", df.shape)

    # 2. Произвести расчет влияния 8 параметров на целевой показатель «Outcome».
    # 3. Дополнительно. Использовать для загрузки данных и расчета библиотеки: Scipy.stats

    # Выделим признаки (8 параметров) и целевую переменную
    features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
    target = 'Outcome'

    if all(col in df.columns for col in features + [target]):
        print(f"\nРасчет влияния 8 параметров на целевой показатель '{target}' (Pearson correlation):")
        print("-------------------------------------------------------------------------")
        for feature in features:
            # Используем scipy.stats.pearsonr для расчета коэффициента корреляции Пирсона и p-значения
            # pearsonr возвращает (коэффициент корреляции, p-значение)
            corr, p_value = stats.pearsonr(df[feature], df[target])
            print(f"Параметр: {feature:<25} | Корреляция: {corr:.4f} | P-значение: {p_value:.4f}")
        print("-------------------------------------------------------------------------")
        print("\nИнтерпретация P-значения:\nЕсли P-значение < 0.05, то связь считается статистически значимой.")
    else:
        print("Ошибка: Не все необходимые столбцы найдены в датасете для расчета корреляции.")
else:
    print(f"Ошибка: Файл {data_file_path} не найден. Убедитесь, что он загружен в среду Colab.")
```
